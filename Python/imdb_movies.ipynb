```python
corr_matrix = df[["year", "rating", "runtime", "votes", "revenue", "metascore"]].corr()
```

```python
plt.figure(figsize = (10, 5))

sns.heatmap(corr_matrix , annot = True)
```

1.   Votes is strongly correlated with revenue in millions (0.61) and ratings (0.51). this means that movies with higher votes tend to earn more revenue and higher rating.
2.   Rating has higher correlation with metascore (0.60), indicating that movies with higher IMDB ratings tend to have higher metascore.
3. Runtime shows moderate correlation with Ratings (0.39) and votes (0.41). this suggests that longer movies tend to get better ratings and more votes
